Функция $plot\_knn\_bound$ принимает на вход объекты $X$, метки классов $y$, метод нормализации признаков $scaler$, число соседей $n\_neighbors$ и границы рисунка $xlim$, $ylim$. Функция обучает $KNN$ классификатор с числом соседей $n\_neighbors$ и визуализирует разделяющую поверхность для классов, полученную с помощью обученного $KNN$ классификатора.

* Если вы не можете различить _выбранные нами цвета_, то измените список цветов в переменных cmap_light и cmap_bold. Актуальный список возможных цветов находится [здесь](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

```python
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets

def plot_knn_bound(X, y, scaler=None, n_neighbors=10, xlim=(-15, 15), ylim=(-20, 20)):
    # step size in the mesh
    h = 0.05

    # Create color maps
    cmap_light = ListedColormap(['C0', 'orange', 'cyan', 'green'][:np.unique(y).shape[0]])
    cmap_bold = ['C0', 'orange', 'c', 'darkgreen'][:np.unique(y).shape[0]]

    x_min, x_max = xlim
    y_min, y_max = ylim
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    grid = np.c_[xx.ravel(), yy.ravel()]

    X_scaled = X # if scaler is None
    if scaler is not None:
        grid = scaler.transform(grid)
        X_scaled = scaler.transform(X)

    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(n_neighbors, algorithm='brute')
    clf.fit(X_scaled, y)

    Z = clf.predict(grid)

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y,
                    palette=cmap_bold, alpha=1.0, edgecolor="black")
    plt.xlabel('x_1')
    plt.ylabel('x_2')
    plt.title('Разделющие поверхности алгоритма {}-NN'.format(n_neighbors))
    plt.grid()
    plt.show()
```

Нарисуем разделющие поверхности $1$-$NN$ и $10$-$NN$

```python
plot_knn_bound(X, y, n_neighbors=1)
plot_knn_bound(X, y, n_neighbors=10)
```

## <font color='DarkOrange'>**Задание 1.1 (кросс-проверка, 1,5 балла)**</font>

 Объясните, почему наблюдается сильное отклоенние разделяющей поверхности от прямой $x_1=0$ при значениях $x_2 < -10$ и $x_2 > 10$. Дайте <font color='OrangeRed'>**строгое математическое обоснование**</font> наблюдаемого явления, показывающее, что в Евклидовом пространстве данная проблема решается нормализацией (скейлингом) признаков.

<font color='MediumOrchid'>**Ваше обоснование тут, можно использовать LaTeX:**</font> \
**Объяснение наблюдаемого явления:**

1. **Сильное отклонение разделяющей поверхности:**

   Отклонение разделяющей поверхности связано с тем, что признаки \( x_1 \) и \( x_2 \) имеют разные диапазоны. При больших абсолютных значениях \( x_2 \) влияние признака \( x_1 \) на расстояние между объектами уменьшается, так как расстояние вычисляется в Евклидовом пространстве:
   
   $$
   d = \sqrt{(x_1^{(i)} - x_1^{(j)})^2 + (x_2^{(i)} - x_2^{(j)})^2}
   $$

   Если диапазон значений \( x_2 \) значительно больше диапазона значений \( x_1 \), то второе слагаемое в этой формуле начинает доминировать, и различия в \( x_1 \) становятся менее значимыми. Это приводит к искривлению разделяющей поверхности, так как расстояние определяется в первую очередь различиями в \( x_2 \), особенно при больших значениях этого признака.

2. **Математическое обоснование:**

   Если \( x_2 \) принимает большие значения по сравнению с \( x_1 \), то, например, при \( x_2 > 10 \),  $$(x_2^{(i)} - x_2^{(j)})^2$$ может быть настолько большим, что разница в \( x_1 \) теряет влияние. Это можно записать в виде:

   $$
   d \approx |x_2^{(i)} - x_2^{(j)}|
   $$

   В этом случае расстояние между объектами будет в основном зависеть от компоненты \( x_2 \), и разделение объектов по \( x_1 \) станет менее значимым. Это и приводит к искривлению разделяющей поверхности в областях, где \( |x_2| \) велико.

3. **Решение через нормализацию (скейлинг):**

   Для решения проблемы необходимо привести все признаки к одному масштабу, чтобы каждый из них одинаково влиял на вычисление расстояния. Обычно это делается с помощью **нормализации (скейлинга)**. Например, стандартный способ нормализации — это приведение каждого признака к нулевому среднему и единичной дисперсии:

   $$
   x'_i = \frac{x_i - \mu_i}{\sigma_i}
   $$

   где $ \mu_i $ — среднее значение признака, а $ \sigma_i $ — стандартное отклонение.

   После нормализации признаки \( x_1 \) и \( x_2 \) будут иметь одинаковый вклад в вычисление расстояний между объектами, что устранит проблему искажения разделяющей поверхности на графике.

 Пусть дано произвольное число точек на плоскости. Представим, что каждая точка - это отдельный класс. Пусть на данной выборке был обучен $1-NN$ классификатор. Чем с геометрической точки зрения являются разделяющие поверхности этого классификатора? Требуется <font color='OrangeRed'>**математически строго, однозначно и с полным обоснованием**</font> определить геометрическое место точек разделяющих поверхностей.

В качестве примера ниже представлена визуализация для четырех точек

```python
X = np.random.randn(4, 2)
y = np.arange(4)
plot_knn_bound(X, y, n_neighbors=1, xlim=(-3, 3), ylim=(-3, 3))
```

<font color='MediumOrchid'>**Ваше обоснование тут, можно использовать LaTeX:**</font>\

1. **Определение задачи:**
   Пусть даны  $N $ точек на плоскости, каждая из которых является отдельным классом. Мы обучаем 1-NN (классификатор ближайших соседей), который для новой точки присваивает тот класс, к чьей точке она ближе в Евклидовом пространстве.

2. **Геометрическое место точек:**
   Пусть $x^{(1)}$, $x^{(2)}$, $\dots$, $x^{(N)}$ — это обучающие точки, каждая из которых принадлежит своему классу. Для того чтобы разделить пространство на области, в каждой из которых ближайшей будет одна из этих точек, строятся **срединные гиперплоскости** (в двумерном случае — прямые). Срединная гиперплоскость между двумя точками $x^{(i)}$ и $ x^{(j)} $ — это множество точек, которые равноудалены от этих двух точек. Она определяется уравнением:

   $$
   \frac{1}{2} \left( \| x - x^{(i)} \|^2 - \| x - x^{(j)} \|^2 \right) = 0
   $$

   Это уравнение можно упростить до:

   $$
   (x - x^{(i)}) \cdot (x^{(i)} - x^{(j)}) = 0
   $$

   Это уравнение описывает гиперплоскость, которая перпендикулярна отрезку, соединяющему $ x^{(i)}$ и $ x^{(j)} $, и проходит через его середину.

3. **Описание разделяющих поверхностей:**
   Разделяющие поверхности в случае 1-NN классификатора являются кусками таких срединных гиперплоскостей между каждой парой точек. Эти гиперплоскости разделяют пространство на области, каждая из которых состоит из точек, ближайших к одной из обучающих точек $ x^{(i)} $. В общем случае для $N $-мерного пространства разделяющими поверхностями являются гиперплоскости размерности $ N-1 $.

4. **Пример для четырех точек на плоскости:**
   В двумерном пространстве, как в приведенном примере, для каждой пары точек строится прямая, которая перпендикулярна отрезку, соединяющему эти две точки, и проходит через середину этого отрезка. Эти прямые являются разделяющими линиями, образующими области, где каждая точка принадлежит классу ближайшей обучающей точки. На графике можно наблюдать такие разделяющие линии.

Таким образом, разделяющие поверхности 1-NN классификатора являются совокупностью срединных гиперплоскостей между каждой парой точек в обучающей выборке.

Как мы видим, объектами обучающей выборки являются тексты. Сырые текстовые данные не могут служить обучающими примерами для алгоритма $KNN$. Перед использованием $KNN$ текстовые данные переменной длины необходимо отобразить в числовой вектор фиксированной размерности.

Работа с текстовыми данными проходит в несколько этапов.

* <font color='plum'>**Токенизация.**</font> Текстовая строка делится на токены. Как правило, в качестве токенов выступают отдельные слова и знаки препинания.

* <font color='plum'>**Нормализация.**</font> Приведение к нижнему регистру, удаление стоп-слов, удаление пунктуации, удаление редких и частотных слов и т.д.

* <font color='plum'>**Векторизация.**</font> Фиксируется словарь токенов $V$ по **обучающей** выборке. Текст представлется в виде вектора длиной $|V|$, где каждый признак есть функция от частоты встречаемости токена в тексте и в обучающей выборке.

В данном задании будет использовано два вида векторизации.

* Первый, $CountVectorizer$, сопоставляет каждому токену частоту встречаемости этого токена в тексте документа.

*  Второй, $TfidfVectorizer$, сопоставляет каждому токену частоту встречаемости этого токена в тексте документа умноженную на величину обратно пропорциональную частоте встречаемости токена во всех документах **обучающей** выборки. Подробне про $tf-idf$ предлагаем изучить самостоятельно, например, [здесь](https://ru.wikipedia.org/wiki/TF-IDF).

<font color='MediumTurquoise'>**Полезные ссылки**</font>

В `sklearn` все векторизаторы автоматически токенизируют текст и приводят его к нижнему регистру. Параметры $max\_df$ и $min\_df$ позволяют убрать из словаря наиболее частые и наиболее редкие токены. Параметр $stop\_words$ позволяет убрать из словаря стоп-слова (малоинформативные слова для данного языка, например, союзы).

```python
import pandas as pd
cnt_accuracies = []
tdif_accuracies = []

for params, score in results.items():
    if params[0] == 'CountVectorizer':
        cnt_accuracies.append(score)
    else:
        tdif_accuracies.append(score)
if cnt_accuracies:
    max_accuracy_cnt = max(cnt_accuracies)
    mean_accuracy_cnt = sum(cnt_accuracies) / len(cnt_accuracies)
else:
    max_accuracy_cnt = None
    mean_accuracy_cnt = None

if tdif_accuracies:
    max_accuracy_tdif = max(tdif_accuracies)
    mean_accuracy_tdif = sum(tdif_accuracies) / len(tdif_accuracies)
else:
    max_accuracy_tdif = None
    mean_accuracy_tdif = None

print("Результаты для CountVectorizer:")
print(f"Максимальная точность: {max_accuracy_cnt}")
print(f"Средняя точность: {mean_accuracy_cnt}")

print("\nРезультаты для TfidfVectorizer:")
print(f"Максимальная точность: {max_accuracy_tdif}")
print(f"Средняя точность: {mean_accuracy_tdif}")

if max_accuracy_cnt is not None and max_accuracy_tdif is not None:
    if max_accuracy_cnt > max_accuracy_tdif:
        print("\nCountVectorizer показывает лучшую максимальную точность.")
    else:
        print("\nTfidfVectorizer показывает лучшую максимальную точность.")

    if mean_accuracy_cnt > mean_accuracy_tdif:
        print("CountVectorizer показывает лучшую среднюю точность.")
    else:
        print("TfidfVectorizer показывает лучшую среднюю точность.")
```

Это может быть связано с тем, что TfidfVectorizer учитывает не только частоту слов (как в CountVectorizer), но и их относительную важность (взвешивание по частоте документа), что позволяет лучше выделять значимые признаки. \
Почему TfidfVectorizer может быть лучше:\

Например, если у вас есть классы, которые расположены близко друг к другу, увеличение K может привести к тому, что модель будет «игнорировать» специфические особенности классов.
